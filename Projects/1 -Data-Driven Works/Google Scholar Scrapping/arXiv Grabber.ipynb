```python
import arxiv
import pandas as pd
import regex as re
from tqdm import tqdm
import glob 
import os
import spacy
def Lexicon_List_Wrapper():
    f = open("lexicon.txt", "r")
    lex_list = []
    for lines in f.readlines():
        if len(lines) > 0:
            title = re.findall(r'\"(.+?)\"', lines)
            lex_list.append(title[0])
    return lex_list

def PaperGrabber(keyword,depth= 10): 
    search = arxiv.Search(query = keyword,  max_results = depth,  sort_by = arxiv.SortCriterion.Relevance)
    data = []
    for result in (search.results()):
        #print(result.title , end ='\n')
        #print(result.summary)
        data.append([result.title ,  result.summary, result.get_short_id(), result.categories, result.published,result.authors])
        #print(data)    #df = pd.DataFrame([data], columns = ['Title', 'Summary','DOI'])
    df = pd.DataFrame(data, columns = ['Title', 'Summary','DOI','Categories','Publish Time','Authors']) 
    return df

def FinalCollector(_from,_to, depth):
    dff=pd.DataFrame()
    for word in tqdm(Lexicon_List_Wrapper()[_from:_to]):
        dff = dff.append(PaperGrabber(word,depth), ignore_index= True)
    filename = "data_from_"+str(_from)+"_to_"+str(_to)+".csv"
    dff.to_csv(filename)
    return dff
    
    
# Merging all the csv files in the one dataframe
def FileMarger():
    files = os.path.join("data_from*.csv")
    files = glob.glob(files)
    df = pd.concat(map(pd.read_csv, files), ignore_index=True)
    df.drop('Unnamed: 0', axis = 1, inplace= True)
    print("Mission Completed  ! ")
    return df
```

```python
nlp=spacy.load("en_core_web_sm")
nlp
```

```python
# Parse text through the `nlp` model
my_text = """The economic situation of the country is on edge , as the stock 
market crashed causing loss of millions. Citizens who had their main investment 
in the share-market are facing a great loss. Many companies might lay off 
thousands of people to reduce labor cost"""

my_doc = nlp(my_text)
type(my_doc)
```

```python
# Printing the tokens of a doc
for token in my_doc:
  print(token.text)
```

```python
for token in my_doc:
  print(token.text,'--',token.is_stop,'---',token.is_punct)
```

```python
# Removing StopWords and punctuations
my_doc_cleaned = [token for token in my_doc if not token.is_stop and not token.is_punct]

for token in my_doc_cleaned:
  print(token.text)
```
